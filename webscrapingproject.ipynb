```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory"
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
# print(soup.prettify())
```

```python
covid_data = soup.find('div', class_='sticky-table-collapsible')
```

```python
headers = [th.text.strip() for th in covid_data.find("tr").find_all("th")[1:]]
print(headers)
```

```python
df = pd.DataFrame(columns = headers)
df
```

```python
rows = []
for row in covid_data.find_all("tr")[1:]:  # Skip the header row
    columns = row.find_all(["td", "th"])  # Include both <td> and <th>
    row_data = [col.text.strip() for col in columns][1:]
    # Ignore empty rows
    if row_data:
        rows.append(row_data)

rows
```

```python
df = pd.DataFrame(rows, columns=headers)
df
```

```python
df.to_csv("covid_data.csv", index=False, encoding="utf-8")
print("\nData successfully saved to covid_data.csv!")
```
